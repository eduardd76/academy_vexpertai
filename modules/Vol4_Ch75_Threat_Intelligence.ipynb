## Actionable CTI for Security Operations

**What You'll Build:**
- CTI feed aggregation (STIX/TAXII)
- IOC management and enrichment
- Threat actor tracking
- TTPs mapping to MITRE ATT&CK

**Sources:**
- Commercial: CrowdStrike, Recorded Future
- Open: AlienVault OTX, Abuse.ch
- OSINT: Twitter, GitHub, dark web

**Impact:** 40% faster threat detection with enriched IOCs

```python
from dataclasses import dataclass
from datetime import datetime
from typing import List, Dict

@dataclass
class IOC:
    ioc_value: str
    ioc_type: str  # ip, domain, hash, url
    threat_type: str
    confidence: float
    source: str
    first_seen: datetime
    last_seen: datetime
    context: Dict

class ThreatIntelPlatform:
    def __init__(self):
        self.iocs = []
    
    def add_ioc(self, ioc: IOC):
        self.iocs.append(ioc)
    
    def enrich_alert(self, ip: str) -> Dict:
        matches = [i for i in self.iocs if i.ioc_value == ip]
        if matches:
            return {
                "threat_score": matches[0].confidence,
                "threat_type": matches[0].threat_type,
                "context": matches[0].context
            }
        return {}

# Example usage
tip = ThreatIntelPlatform()
tip.add_ioc(IOC(
    "185.244.132.77", "ip", "C2", 0.95,
    "CrowdStrike", datetime.now(), datetime.now(),
    {"malware": "Emotet", "campaign": "TA505"}
))

result = tip.enrich_alert("185.244.132.77")
print(f"✅ Threat Intelligence Platform Ready")
print(f"   IOCs loaded: {len(tip.iocs)}")
print(f"   Enrichment: {result}")
```
